### Import hospital dataset

```python
hospitals=pd.read_csv (r'G:\Mi unidad\PUCP\Diplomado de Especialización en Ciencia de Datos para CCSS y Gestión Pública\Python intermedio\Bases\IPRESS.csv', encoding='latin-1')
```

```python
# Filtrar hospitales operativos y con coordenadas válidas
hospitals = hospitals[
    (hospitals["Condición"] == "EN FUNCIONAMIENTO") &
    (~hospitals["NORTE"].isna()) &
    (~hospitals["ESTE"].isna())
]

hospitals_gdf = gpd.GeoDataFrame(
    hospitals,
    geometry=gpd.points_from_xy(
        hospitals["NORTE"],  # X = longitud
        hospitals["ESTE"]    # Y = latitud
    ),
    crs="EPSG:4326"
)
```

```python
print("✅ Hospitales cargados:", hospitals_gdf.shape)
print(hospitals_gdf.geometry.head())
```

### Proximity analysis

```python
# Filtrar Lima y Loreto usando la columna DEP
poblados_lima = poblados[poblados["DEP"] == "LIMA"]
poblados_loreto = poblados[poblados["DEP"] == "LORETO"]

print("✅ Centros poblados en Lima:", poblados_lima.shape[0])
print("✅ Centros poblados en Loreto:", poblados_loreto.shape[0])
print("✅ Hospitales operativos:", hospitals_gdf.shape[0])
```

```python
# 3. Reproyección a UTM (metros)
# =====================
hospitals_utm = hospitals_gdf.to_crs(epsg=32718)   # WGS84 / UTM zone 18S
poblados_lima_utm = poblados_lima.to_crs(epsg=32718)
poblados_loreto_utm = poblados_loreto.to_crs(epsg=32718)

# =====================
# 4. Buffers de 10 km
# =====================
poblados_lima_utm["buffer_10km"] = poblados_lima_utm.geometry.buffer(10000)
poblados_loreto_utm["buffer_10km"] = poblados_loreto_utm.geometry.buffer(10000)
```

```python
# Lima
join_lima = gpd.sjoin(hospitals_utm, 
                      poblados_lima_utm.set_geometry("buffer_10km"), 
                      predicate="within")

hospital_counts_lima = join_lima.groupby("index_right").size()
poblados_lima_utm["hospital_count"] = hospital_counts_lima
poblados_lima_utm["hospital_count"] = poblados_lima_utm["hospital_count"].fillna(0)

# Loreto
join_loreto = gpd.sjoin(hospitals_utm, 
                        poblados_loreto_utm.set_geometry("buffer_10km"), 
                        predicate="within")

hospital_counts_loreto = join_loreto.groupby("index_right").size()
poblados_loreto_utm["hospital_count"] = hospital_counts_loreto
poblados_loreto_utm["hospital_count"] = poblados_loreto_utm["hospital_count"].fillna(0)

# =====================
# 6. Identificar aislado y concentrado
# =====================
# Lima
isolation_lima = poblados_lima_utm.loc[poblados_lima_utm["hospital_count"].idxmin()]
concentration_lima = poblados_lima_utm.loc[poblados_lima_utm["hospital_count"].idxmax()]

# Loreto
isolation_loreto = poblados_loreto_utm.loc[poblados_loreto_utm["hospital_count"].idxmin()]
concentration_loreto = poblados_loreto_utm.loc[poblados_loreto_utm["hospital_count"].idxmax()]

# =====================
# 7. Mostrar resultados
# =====================
print("\n👉 Resultados Lima")
print("Más aislado:", isolation_lima["NOM_POBLAD"], "-", isolation_lima["hospital_count"], "hospitales")
print("Más concentrado:", concentration_lima["NOM_POBLAD"], "-", concentration_lima["hospital_count"], "hospitales")

print("\n👉 Resultados Loreto")
print("Más aislado:", isolation_loreto["NOM_POBLAD"], "-", isolation_loreto["hospital_count"], "hospitales")
print("Más concentrado:", concentration_loreto["NOM_POBLAD"], "-", concentration_loreto["hospital_count"], "hospitales")
```

### Proximity Analysis using Folium

```python
# --- Reproyectar puntos y buffers a EPSG:4326 (para Folium)
iso_buffer = gpd.GeoDataFrame(
    geometry=[isolation_lima.geometry.buffer(10000)], crs="EPSG:32718"
).to_crs(epsg=4326)

conc_buffer = gpd.GeoDataFrame(
    geometry=[concentration_lima.geometry.buffer(10000)], crs="EPSG:32718"
).to_crs(epsg=4326)

iso_point = gpd.GeoDataFrame(
    geometry=[isolation_lima.geometry], crs="EPSG:32718"
).to_crs(epsg=4326)

conc_point = gpd.GeoDataFrame(
    geometry=[concentration_lima.geometry], crs="EPSG:32718"
).to_crs(epsg=4326)

# --- Crear mapa centrado en Lima
map_lima = folium.Map(location=[-12.0464, -77.0428], zoom_start=8)

# 🔴 Buffer aislado
folium.GeoJson(
    iso_buffer.geometry.iloc[0],
    style_function=lambda x: {"color": "red", "fillColor": "red", "fillOpacity": 0.2}
).add_to(map_lima)

folium.Marker(
    location=[iso_point.geometry.iloc[0].y, iso_point.geometry.iloc[0].x],
    popup=f"<b>Aislado:</b> {isolation_lima['NOM_POBLAD']}<br>Hospitales: {int(isolation_lima['hospital_count'])}",
    tooltip="Centro poblado más aislado",
    icon=folium.Icon(color="red")
).add_to(map_lima)

# 🟢 Buffer concentrado
folium.GeoJson(
    conc_buffer.geometry.iloc[0],
    style_function=lambda x: {"color": "green", "fillColor": "green", "fillOpacity": 0.2}
).add_to(map_lima)

folium.Marker(
    location=[conc_point.geometry.iloc[0].y, conc_point.geometry.iloc[0].x],
    popup=f"<b>Concentrado:</b> {concentration_lima['NOM_POBLAD']}<br>Hospitales: {int(concentration_lima['hospital_count'])}",
    tooltip="Centro poblado más concentrado",
    icon=folium.Icon(color="green")
).add_to(map_lima)

map_lima
```
